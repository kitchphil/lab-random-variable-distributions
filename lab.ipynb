```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
df
```

**Get the numerical variables from our dataset.**

```python
numerical = df.select_dtypes(include=np.number)
```

```python
numerical.dtypes
```

**Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.**

```python
for i in numerical.columns:
    sns.displot(numerical, x=i, kde=True)
```

**Check if any of the transformations we have seen up to this point changes the result.**

**Log-Transform**

```python
nums = np.log(numerical)

for i in nums.columns:
    sns.displot(nums, x=i, kde=True)
```

as the outliers have been removed using the log-transform, the graphs change, the scale is reduced. None of the changes are significant enough, however, to warrant further investigation.

**Standard Scaler & MinMax Scaler**

```python
numsta = StandardScaler().fit(numerical)
num_standardized = numsta.transform(numerical)
numstd = pd.DataFrame(num_standardized, columns=numerical.columns)

for i in numstd.columns:
    sns.displot(numstd, x=i, kde=True)
```

```python
numinmax = MinMaxScaler().fit(numerical)
num_normalized = numinmax.transform(numerical)
numnx = pd.DataFrame(num_normalized, columns=numerical.columns)

for i in numnx.columns:
    sns.displot(numnx, x=i, kde=True)
```

as these scalers essentially do the same thing (except reduce the axes from 0 to 1 for the MinMax, and Adjust the values to be centered around 0 for Standard Scaler), we can conclude that they will not change anything between each other, nor the original dataframe.
